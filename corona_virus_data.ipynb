# 6) Pivot Tables

# 7) Crosstab


- Table showing the relationship between two or more variables. Where the table only shows the relationship between two categorical variables, a crosstab
is also known as a contingency table.
- It helps us to make informed decisions regarding your research by identifying patterns, trends, and the correlation between your study parameters.


```python
pd.crosstab(cd["WHO Region"],cd["Country"])
```

Cross Tab is show the observation country by WHO Region.

# 8) Total Counting by Country

```python
Total = cd.groupby("WHO Region")["Deaths"].sum()
Total
```

```python
Death_Case = cd[['Country','Deaths']].sort_values(by=['Deaths'],ascending=False).head(10)
Death_Case
```

```python
Recovery_Case = cd[['Country','Recovered']].sort_values(by=['Recovered'],ascending=False).head(10)
Recovery_Case
```

```python
Active_Case = cd[['Country','Active']].sort_values(by=['Active'],ascending=False).head(10)
Active_Case
```

```python
cd.groupby(['WHO Region'])[['Country']].count()
```

- The count of Country Group according to WHO Region.

# Visualisation

1) Bar Plot

```python
plt.figure(figsize=(18,5))
plt.bar('WHO Region','Deaths',data=cd,color='g')
plt.xticks(rotation=0)
plt.show()
```

2) Recovery Rate Plot


```python
plt.figure(figsize=(18,5))
plt.bar('WHO Region','Recovered',data=cd,color='r')
plt.xticks(rotation=0)
plt.show()
```

3) Active Case Plot


```python
plt.figure(figsize=(18,5))
plt.bar('WHO Region','Active',data=cd,color='cyan')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x=cd['WHO Region'],palette='coolwarm_r')
plt.xticks(rotation=0)
plt.show()
```

# 11) Heat Map


```python
sns.heatmap(cd.corr())
plt.show()
```

# 12) PIE Plot

```python
plt.title('WHO Region',fontsize=20)
cd['WHO Region'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```
